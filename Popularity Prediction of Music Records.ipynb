### Comment: 1990 and 1991 are the 2 years having the highest frequency of top10 songs => 1990 and 1991 can be considered the most successful time during given period. The other years only produced output as half as the first years did.

### We'll divide the time periods into three periods.
#### `Period - 1: <=1991`
#### `Period - 2: 1992-1997`
#### `Period - 3: >1997`

```python
#Evaluate the correlation between top10 songs and elements that create a song through 3 periods

period1=df[df['year']<=1991]
period2=df[(df['year']<=1997) & (df['year']>=1992)]
period3=df[df['year']>1997]
```

```python
period_table=pd.DataFrame([period1.mean(),period2.mean(),period3.mean()],index=['period1','period2','period3'])
period_table['Top10_Avg']=[len(period1[period1.Top10==1])/2,len(period2[period2.Top10==1])/6,len(period3[period3.Top10==1])/14]
period_table=period_table.drop('year',axis=1)
```

#### Highly Imbalanced Dataset so we'll be performing oversampling on Minority Class

```python
features=df_final.drop('Top10',axis=1)
target=df_final[['Top10']]
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
os=RandomOverSampler(random_state=40)
```

```python
features,target=os.fit_resample(features,target)
```

```python
target.value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
features=pd.DataFrame(scaler.fit_transform(features),index=features.index,columns=features.columns)
```

```python
from sklearn.model_selection import train_test_split,cross_val_score,KFold,GridSearchCV,RandomizedSearchCV
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(features,target,train_size=0.8,random_state=40)
xval,xtest,yval,ytest=train_test_split(xtest,ytest,train_size=0.5,random_state=40)
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import MultinomialNB
```

### Cross Validation

```python
score_list={'SVC()':0,'RandomForestClassifier()':0,'GradientBoostingClassifier()':0,"MultinomialNB()":0}
```

```python
for model in [SVC(),RandomForestClassifier(),GradientBoostingClassifier(),MultinomialNB()]:
    model=model
    model.fit(xtrain,ytrain)
    score=cross_val_score(model,xtrain,ytrain,cv=KFold(n_splits=5,random_state=70,shuffle=True)).mean()
    score_list[str(model)]=np.round(score*100,3)
```

```python
score_list
```

### Random Forest

```python
modelrf=RandomForestClassifier(n_jobs=-1)
param={'n_estimators':[100,200,300],
    'criterion':["gini", "entropy"],
    'max_depth':[20,30,40],
    'random_state':[30,40,50,60]}

gridcv=GridSearchCV(modelrf,param_grid=param,cv=KFold(n_splits=5,random_state=70,shuffle=True),n_jobs=-1,verbose=2)
```

```python
gridcv.fit(xval,yval)
```

```python
gridcv.best_score_
```

```python
gridcv.best_estimator_
```

```python
modelrf=RandomForestClassifier(max_depth=40,random_state=40,n_estimators=200)
modelrf.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,brier_score_loss
```

```python
ypred=modelrf.predict(xval)
confusion_matrix(yval,ypred)
```

```python
print(classification_report(yval,ypred))
```

```python
modelrf.score(xval,yval)
```

```python
pred_prob=modelrf.predict_proba(xval)[:,1]
```

```python
threshold=np.arange(0,1.05,0.05)
tpr,fpr=[],[]
for thre in threshold:
    res=[np.where(x>thre,1,0) for x in pred_prob]
    cm=confusion_matrix(yval,res)
    tpr.append(cm[1,1]/(cm[1,0]+cm[1,1]))
    fpr.append(cm[0,1]/(cm[0,0]+cm[0,1]))
```

```python
plt.plot(fpr,tpr)
plt.plot(np.arange(0,1,1/757),np.arange(0,1,1/757),linestyle='dashed',c='r')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
```

```python
#Threshold selected = 0.65
ypred=[np.where(x>0.65,1,0) for x in pred_prob]
```

```python
confusion_matrix(yval,ypred)
```

```python
print(classification_report(yval,ypred))
```

### Final Prediction on Test Data

```python
#Threshold selected = 0.65
pred_prob=modelrf.predict_proba(xtest)[:,1]
ypred=[np.where(x>0.65,1,0) for x in pred_prob]
```
